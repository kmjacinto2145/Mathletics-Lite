```python
import numpy as np
import pandas as pd
import time
import os
```

```python
def clear():
    os.system( 'cls' )
```

```python
question_count = 20

operations = {1: ["+"], 2: ["+","-"], 3: ["+","-","*"], 4: ["+","-","*", "/"], 5: ["+","-","*", "/"]}

add_number_ranges = {1: (0,10), 2: (0,20), 3: (0,50), 4: (0,99), 5: (-99,99)}
subtract_number_ranges = {2: (0,10), 3: (0,20), 4: (-10,50), 5: (-50,99)}
multiply_number_ranges = {3: (0,10), 4: (-5,15), 5: (-10,20)}
divide_number_ranges = {4: (1,100), 5: (-10,200)}
```

```python
def add(number_range, answers_correct):
    
    #computer calculations before player input
    numbers = np.random.randint(number_range[0], number_range[1], size = 2)

    print(numbers[0], "+", numbers[1], "=", "?")
    result = numbers[0] + numbers[1]
    
    #ask for player input
    answer = int(input())
    
    #checks player's answer
    if result == answer and answer == 69:
        print("Nice!")
        answers_correct += 1
    elif result == answer:
        print("Correct!")
        answers_correct += 1        
    else:
        print("Wrong!")   
    
    return answers_correct
```

```python
def subtract(number_range, answers_correct):
    
    #computer calculations before player input
    numbers = np.random.randint(number_range[0], number_range[1], size = 2)

    print(numbers[0], "-", numbers[1], "=", "?")
    result = numbers[0] - numbers[1]
    
    #ask for player input
    answer = int(input())
    
    #checks player's answer
    if result == answer and answer == 69:
        print("Nice!")
        answers_correct += 1
    elif result == answer:
        print("Correct!")
        answers_correct += 1     
    else:
        print("Wrong!")   
    
    return answers_correct 
```

```python
def multiply(number_range, answers_correct):
    
    #computer calculations before player input
    numbers = np.random.randint(number_range[0], number_range[1], size = 2)

    print(numbers[0], "*", numbers[1], "=", "?")
    result = numbers[0] * numbers[1]
    
    #ask for player input
    answer = int(input())
    
    #checks player's answer
    if result == answer and answer == 69:
        print("Nice!")
        answers_correct += 1
    elif result == answer:
        print("Correct!")
        answers_correct += 1     
    else:
        print("Wrong!")   
    
    return answers_correct 
```

```python
def divide(number_range, answers_correct):
    
    valid = False
    while not valid:
        
        numbers = np.random.randint(number_range[0], number_range[1], size = 2)
        
        #check for division by zero
        if numbers[1] == 0:
            continue
        if numbers[0] % numbers[1] == 0:
            valid = True

    #computer calculations before player input
    print(numbers[0], "/", numbers[1], "=", "?")
    result = numbers[0] / numbers[1]
    
    #ask for player input
    answer = int(input())
    
    #checks player's answer
    if result == answer and answer == 69:
        print("Nice!")
        answers_correct += 1
    elif result == answer:
        print("Correct!")
        answers_correct += 1     
    else:
        print("Wrong!")   
    
    return answers_correct     
```

```python
def question(level, answers_correct):
    
    global operations
    operation = np.random.choice(operations[level])
    
    if operation == "+":
        
        global add_number_ranges
        answers_correct = add(add_number_ranges[level], answers_correct)
    
    elif operation == "-":
        
        global subtract_number_ranges
        answers_correct = subtract(subtract_number_ranges[level], answers_correct)
    
    elif operation == "*":
    
        global multiply_number_ranges
        answers_correct = multiply(multiply_number_ranges[level], answers_correct)
    
    elif operation == "/":
        
        global divide_number_ranges
        answers_correct = divide(divide_number_ranges[level], answers_correct)
    
    clear()
    
    return answers_correct
```

```python
def play_round(level):
    
    #Starts the clock
    start_time = time.time()
    
    global question_count
    
    print("Questions to complete:", question_count)
    print("Start!")
    answers_correct = 0
    
    
    for i in range(question_count):
        print("Question", i + 1)
        prev_answers_correct = answers_correct
        answers_correct = question(level, answers_correct)
        
        if answers_correct == prev_answers_correct:
            print("Game over!")
            return
            
    end_time = time.time()
    
    #Calculates time elapsed
    time_elapsed = end_time - start_time
    
    #Results for player
    print("Time elapsed:", round(time_elapsed, 3), "seconds")
    print("Time per question:", round(time_elapsed / question_count, 3), "seconds")
```

```python
def main():
    level = int(input("What level would you like? "))
    play_round(level)            
```

```python
if __name__ == "__main__":
    main()
```
